```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
```

```python
#selenium으로 웹 드라이버 실행
service = Service(executable_path=ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

#url 전달
url = 'https://news.kbs.co.kr/news/pc/category/category.do?ref=pSiteMap#20240310&1'
driver.get(url)

#기다려달라는 값을 전달: 최대 10초 기다리기
wait = WebDriverWait(driver, 10)

#드라이버에 접근 > 페이지 소스 가져오기
html = driver.page_source

#드라이버 종료
driver.quit()
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
box_contents = soup.find_all('a', class_='box-content flex-style')
box_contents
```

```python
#원하는 부분들만 찾아서, 리스트에 담고, 데이터 프레임으로 만들기 작업
#빈 리스트 만들기
url_list = []
title_list = []
body_list = []
date_list = []

for box_content in box_contents:
    url = 'https://news.kbs.co.kr/' + box_content['href']
    title = box_content.find('p', class_='title').text
    body = box_content.find('p', class_='news-text').text
    date = box_content.find('span', class_='date').text
    
    #빈 리스트에 값 append(붙이기)
    url_list.append(url)
    title_list.append(title)
    body_list.append(body)
    date_list.append(date)
```

```python
#데이터 저장
data = {'뉴스 url':url_list, '제목':title_list, '내용':body_list, '날짜':date_list}
df = pd.DataFrame(data)
df
```

```python
#페이지네이션: 1~13페이지를 돌면서 크롤링
#for문 f string
for i in range(1,14):
    print(f"{i}번째 페이지")
```

```python
#빈 리스트 만들기
url_list = []
title_list = []
body_list = []
date_list = []

#서비스 변수 설정
service = Service(executable_path=ChromeDriverManager().install())

#1. 페이지를 돌면서 작업을 수행
for page_num in range(1,14):
    #selenium으로 웹 드라이버 실행
    driver = webdriver.Chrome(service=service)

    #url 전달
    url = f'https://news.kbs.co.kr/news/pc/category/category.do?ref=pSiteMap#20240310&{page_num}'
    driver.get(url)

    #기다려달라는 값을 전달: 최대 10초 기다리기
    wait = WebDriverWait(driver, 10)

    #드라이버에 접근 > 페이지 소스 가져오기
    html = driver.page_source

    #드라이버 종료
    driver.quit()
    
    #2.BeautifrlSoup 사용해서 파싱 > 태그들을 찾아서 담아줌
    soup = BeautifulSoup(html, "html.parser")
    box_contents = soup.find_all("a", class_="box-content flex-style")

    #3. 박스 데이터 돌면서 리스트에 데이터 담기
    for box_content in box_contents:
        url = 'https://news.kbs.co.kr/' + box_content['href']
        title = box_content.find('p', class_='title').text
        body = box_content.find('p', class_='news-text').text
        date = box_content.find('span', class_='date').text
        
        #빈 리스트에 값 append(붙이기)
        url_list.append(url)
        title_list.append(title)
        body_list.append(body)
        date_list.append(date)

#4. 데이터 프레임으로 저장
data = {'뉴스 url':url_list, '제목':title_list, '내용':body_list, '날짜':date_list}
df = pd.DataFrame(data)

#5. csv 파일로 저장
df.to_csv('news_kbs_pagenation.csv', index=False)
```
