```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#한글 폰트 설정
plt.rc('font', family='Malgun Gothic')
```

```python
df = pd.read_csv('tour_review_backup.csv')
df
```

```python
#1. 카테고리별 리뷰 시각화
#각 카테고리(가족여행, 친구모임, 커플여행 등) 별로 리뷰 수를 시각화하고, 어떤 유형이 가장 많은지 체크

#카테고리별 리뷰 수 계산
category_count = df['category'].value_counts()
category_count
```

```python
plt.figure(figsize=(15, 6))
category_count.plot(kind='bar',color = 'skyblue')

plt.title("카테고리별 리뷰 수")
plt.xlabel("카테고리")
plt.ylabel("리뷰 수")
plt.xticks(rotation=20)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#2. 날짜별 리뷰 분포 시각화
#특정 날짜에 리뷰가 많거나 적은 때를 확인
#날짜를 datetime 형식으로 변환

import matplotlib.dates as mdates

df['date'] = pd.to_datetime(df['date'])
```

```python
#날짜별 리뷰 수 계산 1탄
date_counts = df['date'].value_counts().sort_index()

plt.figure(figsize=(30, 10))
sns.lineplot(x=date_counts.index, y=date_counts.values, marker='o', color='orange')

plt.title("날짜별 리뷰 수")
plt.xlabel("날짜")
plt.ylabel("리뷰 수")

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#날짜별 리뷰 수 계산 2탄
date_counts = df['date'].value_counts().sort_index()

plt.figure(figsize=(30, 10))
sns.lineplot(x=date_counts.index, y=date_counts.values, marker='o', color='orange')

plt.title("날짜별 리뷰 수")
plt.xlabel("날짜")
plt.ylabel("리뷰 수")

#xticks에 날짜 형식 설정
date_format = mdates.DateFormatter('%Y-%m')
plt.gca().xaxis.set_major_formatter(date_format)
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=50)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#3. 나이별 평균 평점 비교
#각 연령대별로 평균 평점 비교 > 어떤 연령대가 평균적으로 좋은 리뷰를 남기는지 분석

#나이별 평균 평점 계산
age_rating = df.groupby('age')['rating'].mean().sort_values()

#시각화
plt.figure(figsize=(10, 6))
sns.lineplot(x=age_rating.index, y=age_rating.values, palette='viridis')

plt.title("나이별 평균 평점")
plt.xlabel("나이")
plt.ylabel("평균 평점")

plt.xticks(rotation=50)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#낮은 순 정렬

df_age = df[df['age'].str.endswith('대')]
df_age = df_age[df_age['rating'].notnull()]

#나이별 평균 평점 계산
age_rating_new = df_age.groupby('age')['rating'].mean().sort_values()

#시각화
plt.figure(figsize=(10, 6))
sns.barplot(x=age_rating_new.index, y=age_rating_new.values, palette='viridis')

plt.title("나이별 평균 평점")
plt.xlabel("나이")
plt.ylabel("평균 평점")

#y축 범위 설정
plt.ylim(4, None)

plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#나이순 정렬

age_rating_new = age_rating_new.sort_index()

#시각화
plt.figure(figsize=(10, 6))
sns.barplot(x=age_rating_new.index, y=age_rating_new.values, palette='viridis')

plt.title("나이별 평균 평점")
plt.xlabel("나이")
plt.ylabel("평균 평점")

#y축 범위 설정
plt.ylim(4, None)

plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
