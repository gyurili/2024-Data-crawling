1. RSS란? -웹사이트에서 컨텐츠를 쉽게 배포하기 위한 표준 포맷
   -RSS: 사용자가 관심있는 웹사이트의 업데이트를 받아보고자 할 때 사용 -크롤링: 웹사이트의 데이터를 자동으로 수집하고 분석하기 위해 사용 -둘 다 웹 콘텐츠를 가져오기 위해 사용되는 방법

2. 파이썬으로 RSS 사용해서 뉴스 데이터 크롤링하기

```python
import feedparser
import pandas as pd
```

```python
url = 'https://news.sbs.co.kr/news/SectionRssFeed.do?sectionId=07&plink=RSSREADER'
feed = feedparser.parse(url)
feed
```

```python
#피드 항목들을 순회하며 정보 출력
for entry in feed.entries:
    print("\ntitle", entry.title)
    print("url", entry.link)
    print("Date", entry.published)
    print("Summary", entry.summary)
    
```

```python
#피드 제목 출력
print("Feed title:", feed.feed.title)
```

```python
# 빈 리스트 만들기
url_list = []
title_list = []
summary_list = []
date_list = []

# feed entries 순회하면서 리스트에 값 추가
for entry in feed.entries:
    url_list.append(entry.link)
    title_list.append(entry.title)
    summary_list.append(entry.summary)
    date_list.append(entry.published)
```

```python
data = {'뉴스 url':url_list, '제목':title_list, '내용 요약':summary_list, '날짜':date_list}
df = pd.DataFrame(data)
df.to_csv('news_sbs_rss.csv')
```
